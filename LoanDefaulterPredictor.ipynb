```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
#load the dataset
df = pd.read_csv("train.csv")
pd.set_option("display.max_rows", 10, "display.max_columns", None)
df
```

```python
#check the column names
print(df.columns.unique())
#check the number of unique values in each column
print("\n")
for i in range(len(df.columns)) :
    print("Column ",i," is ",df.columns[i], ".It has a data type of ", df[df.columns[i]].dtypes)
    print("It has ", len(df[df.columns[i]].unique()), " unique values.\n")
```

```python
# Remove CITY and Id column because it will cause trouble during one hot encoding.
df = df.drop(['Id', 'CITY'], axis = 1)
#Check for missing values
print(df.isnull().sum())
df
```

```python
#one hot encode Married/Single , House_Ownership , Car_Ownership , STATE , Profession
df = pd.get_dummies(df , columns = ["Married/Single" , "House_Ownership" , "Car_Ownership" , "STATE" , "Profession"])
df
```

```python
X = df.drop('Risk_Flag', axis = 1)
y = df['Risk_Flag']
#split the data into train and test data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = RandomForestClassifier(n_jobs= - 1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```
